# Snowboarding Insights
The snowboard dataset (snowboard.csv) was obtained from: https://www.kaggle.com/datasets/oznbgcvn/snowboardpowdervalue

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
snowboard_data = pd.read_csv('snowboard.csv')
snowboard_data.info()
```

## Snowboard Brands

```python
brands = snowboard_data['Brand'].unique()
brands
```

*Note the whitespace in some of the brand names, which needs to be removed.*

```python
snowboard_data['Brand'] = snowboard_data['Brand'].str.strip() # remove whitespace

unique_brands = snowboard_data['Brand'].unique()
unique_brands
```

```python
brand_counts = snowboard_data['Brand'].value_counts()
brand_counts
```

This project focuses on the following brands:
- Arbor
- Burton
- Capita
- GNU
- Jones
- Lib Tech
- K2
- Rome
- Salomon

*These brands were selected by the analyst based on snowboard brands of interest.*

```python
brands_of_interest= ['Arbor', 'Burton', 'Capita', 'Gnu', 'Jones',
                     'Lib Tech', 'K2', 'Rome', 'Salomon']

snowboards = snowboard_data[snowboard_data['Brand'].isin(brands_of_interest)]
```

### Types of Riding Styles

```python
riding_styles = snowboards['Riding Style'].unique()
riding_styles
```

This project focuses on the following riding styles:
- All Mountain
- Alternative Freeride
- Freeride
- Powder
- Snowsurf

*These brands were selected by the analyst based on interest in specific riding styles.*

```python
riding_styles_of_interest = ['All Mountain', 'Alternative Freeride', 'Freeride',
                             'Powder', 'Snowsurf']

snowboards = snowboards[snowboards['Riding Style'].isin(riding_styles_of_interest)]
```

```python
snowboards.head()
```

-----

## All Mountain Snowboards

```python
all_mountain_boards = snowboards[snowboards['Riding Style'] == 'All Mountain']
all_mountain_boards
```

```python
all_mountain_shapes = pd.DataFrame(all_mountain_boards['Shape'].value_counts())
all_mountain_shapes
```

```python
plt.figure(figsize=(10,6))
all_mountain_shapes.plot(kind='bar', color='#0b9473')
plt.title('All Mountain Snowboard Shapes')
plt.xlabel('')
plt.ylabel('Number of Boards')
```

Many mountain boards are directional and designed with a twin shape.

```python
all_mountain_camber = all_mountain_boards['Camber Profile'].value_counts()
all_mountain_camber
```

Majority of the all mountain boards are camber.

```python
all_mountain_flex = all_mountain_boards['Flex'].value_counts()
all_mountain_flex
```

All mountain boards have a medium flex.
